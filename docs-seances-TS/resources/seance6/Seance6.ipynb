```python
def plot_up_to(g,b,N,func_name=None,step=1):
    
    fig = plt.figure(0, figsize=(8,6), dpi=100)
    
    
    
    if func_name is None:
        fig.suptitle(f"Intégrales successives d'une suite de fonctions")
    else:
        fig.suptitle(f"Intégrales successives de la suite de fonctions {func_name}")
    
        
    ax = fig.add_subplot(111)
    
    ax.grid(True)
    
    xrange = np.linspace(0,b,200)
    
    for i in range(1,N):
        yvals = g(xrange,step*i)
        ax.plot(xrange,yvals)
```

```python
def g(x,n):
    return np.log(1+x)*np.sin(0.001*n*x)

plot_up_to(g,10,50,r"$x\longmapsto e^{-nx}$",10)
```
