```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
mpl.rc('font',**{'family':'Times New Roman'})
mpl.rc('text',usetex=True)
```

```python
np.random.rand()
```

```python
def monkeys(p,Tmax):
    Xt = [0]
    for _ in range(Tmax):
        if np.random.rand() <= p:
            Dx = -1
        else:
            Dx = 1
        Xt.append(Xt[-1] + Dx)
    return Xt
```

```python
def graphMonkeys(p,Tmax):
    Xt = monkeys(p,Tmax)
    fig = plt.figure(0,figsize=(6,4),dpi=100)
    ax = fig.add_subplot(111)
    ax.set_xlabel('$t$')
    ax.set_ylabel('$X_t$')
    ax.grid(True)
    ax.plot(Xt,'b')
    fig.show()
    return fig
```

```python
%matplotlib inline
```

```python
fig = graphMonkeys(0.5,200)
```
