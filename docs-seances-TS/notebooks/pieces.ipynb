```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def probleme_piece(N, show=True):
    
    import time
    start = time.time()
    
    # créer une tas de N pièces avec des masses identiques
    masse = 1.0
    X = np.array([masse for _ in range(N)])
    
    # on choisit un indice à changer, au hasard
    indices = np.arange(0,N)
    diff_index = np.random.choice(indices)
    X[diff_index] *= 0.8
    
    if show:
        print(X)
    
    
    def solve(X, index):
        N = X.size
        if N<2:
            return index
        if N==2:
            if X[0] < X[1]:
                return index
            else:
                return index+1
        
        tas = np.split(X, [N//3, 2*N//3])
        
        if tas[0].sum() < tas[1].sum():
            return solve(tas[0], index)
        elif tas[1].sum() < tas[0].sum():
            return solve(tas[1], index+N//3)
        else:
            return solve(tas[2], index+2*N//3)
    
    end = time.time()
    
    return solve(X, 0), (end-start)
        
```

```python
probleme_piece(10)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
output_notebook
```

```python
def exec_time(Nmin, Nmax):
    
    results = []
    
    xrange = np.arange(Nmin, Nmax)
    
    for i in range(Nmin, Nmax):
        duration = 0
        
        for _ in range(50):
            _, extime = probleme_piece(i, False)
            duration += extime
        
        results.append(duration)
    
    results = np.array(results)
    
    def smoothie(X, k):
        N = X.size
        
        for i in range(k, N):
            X[i] = X[i-k:k].sum()/k

    smoothie(results, 1)
    
    p = figure()
    p.line(xrange, results)
    
    show(p)
    
```

```python
exec_time(10000, 20000)
```
